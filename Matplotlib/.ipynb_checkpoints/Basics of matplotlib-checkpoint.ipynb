```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,3,12)
y = x ** 4
```

```python
fig = plt.figure()


axis = fig.add_axes([0.1,0.1,1.0,1.0])

axis.plot(x,y)
axis.set_xlabel('X')
axis.set_ylabel('Y')
axis.set_title('Basics')
```

```python
fig, axis = plt.subplots(nrows=1,ncols=2)

axis[0].plot(x,y)
axis[0].set_title('First')

axis[1].plot(y,x)
axis[1].set_title('Second')
```

```python
plt.scatter(x,y)
```
